```python
independent=dataset[["cement","slag",'flyash', 'water', 'superplasticizer',
       'coarseaggregate', 'fineaggregate', 'age']]#
independent
```

```python
dependent=dataset[["csMPa"]]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.01,random_state=10)
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
weight=regressor.coef_
weight
bais=regressor.intercept_
bais
y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle
filename="/content/drive/MyDrive/Colab Notebooks/Machine Learning-Classification/Assignment/1. AI in Civil/simple_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
loaded_model=pickle.load(open("/content/drive/MyDrive/Colab Notebooks/Machine Learning-Classification/Assignment/1. AI in Civil/simple_linear.sav",'rb'))
result=loaded_model.predict([[540.0,	0.0,	0.0	,162.0,	2.5	,1040.0	,676.0,	90]])
result
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.01 ,random_state=200)
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators =  190, random_state = 200)
regressor.fit(X_train, y_train)

y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
# highest value in random forest model is 0.98251
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

from sklearn.svm import SVR
regressor=SVR(kernel="poly",C=100.0,gamma="auto")

regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
